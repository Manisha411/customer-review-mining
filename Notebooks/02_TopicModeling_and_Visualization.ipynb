```python
import pandas as pd
df = pd.read_csv("data/beauty_reviews_filtered.csv", low_memory=False)
df.head(2)
```

```python
import re

def clean_text(t):
    t = str(t)
    t = re.sub(r'http\S+|www\S+', ' ', t)         # remove URLs
    t = re.sub(r'[^A-Za-z0-9\s]', ' ', t)         # keep alphanumerics
    t = re.sub(r'\s+', ' ', t).strip()
    return t.lower()

df = df.dropna(subset=['reviews.text']).copy()
df['clean_text'] = df['reviews.text'].map(clean_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Vectorize
vectorizer = CountVectorizer(stop_words='english', max_df=0.9, min_df=10)  # if data is small: min_df=5 or 3
X = vectorizer.fit_transform(df['clean_text'])

# Fit LDA
n_topics = 6   # try 5–8 and pick the clearest
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
doc_topic = lda.fit_transform(X)

# Inspect top words per topic
def show_topics(model, feature_names, topn=12):
    for k, comp in enumerate(model.components_):
        top_idx = comp.argsort()[-topn:][::-1]
        words = [feature_names[i] for i in top_idx]
        print(f"Topic {k}: {', '.join(words)}")

feature_names = vectorizer.get_feature_names_out()
show_topics(lda, feature_names, topn=12)
```

```python
df['topic_id'] = doc_topic.argmax(axis=1)
df[['reviews.rating','reviews.text','topic_id']].head(3)
```

```python
# After reading top words and sample reviews per topic, define labels:
topic_labels = {
    0: "Hydration/Moisture",
    1: "Irritation/Sensitivity",
    2: "Longevity/Smudge-Proof",
    3: "Texture/Absorption",
    4: "Scent/Fragrance",
    5: "Packaging/Delivery"
}
df['topic_label'] = df['topic_id'].map(topic_labels)
```

```python
df.to_csv("data/beauty_reviews_with_topics.csv", index=False)
```

```python
import numpy as np

# Ensure numeric ratings if present
if 'reviews.rating' in df.columns:
    df['reviews.rating'] = pd.to_numeric(df['reviews.rating'], errors='coerce')

# If you created sentiment earlier (positive/neutral/negative), keep using it.
# If not, you can add a quick label from stars as a placeholder:
def star_to_sentiment(r):
    if pd.isna(r): return None
    r = float(r)
    if r >= 4: return 'positive'
    if r <= 2: return 'negative'
    return 'neutral'

if 'sentiment' not in df.columns:
    df['sentiment'] = df['reviews.rating'].map(star_to_sentiment)

# Brand-level summary
brand_col = 'brand' if 'brand' in df.columns else 'name'  # fallback
agg_brand = (
    df.groupby(brand_col)
      .agg(
          n_reviews=('reviews.text','count'),
          avg_rating=('reviews.rating','mean'),
          pos_rate=('sentiment', lambda s: (s=='positive').mean() if s.notna().any() else np.nan),
          neg_rate=('sentiment', lambda s: (s=='negative').mean() if s.notna().any() else np.nan)
      )
      .sort_values('n_reviews', ascending=False)
      .reset_index()
)

# Topic counts by brand
agg_topic = (
    df.groupby([brand_col, 'topic_label' if 'topic_label' in df.columns else 'topic_id'])
      .size()
      .reset_index(name='count')
      .sort_values(['count'], ascending=False)
)

agg_brand.to_csv("data/agg_brand.csv", index=False)
agg_topic.to_csv("data/agg_topic.csv", index=False)

agg_brand.head(10), agg_topic.head(10)
```

```python
import matplotlib.pyplot as plt

# Rating distribution
if 'reviews.rating' in df.columns:
    df['reviews.rating'].dropna().astype(int).value_counts().sort_index().plot(kind='bar', title='Rating Distribution')
    plt.show()

# Sentiment share
if 'sentiment' in df.columns:
    df['sentiment'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Sentiment Distribution')
    plt.ylabel('')
    plt.show()

# Top topics overall
df['topic_label' if 'topic_label' in df.columns else 'topic_id'].value_counts().head(10).plot(kind='bar', title='Top Topics')
plt.show()
```
